```python
# Activity 1
import numpy as np
temperature = np.array([36.5, 37.0, 37.5, 38.0, 38.5, 39.0, 39.5])

highest = np.max(temperature)
lowest = np.min(temperature)
average = np.mean(temperature)

print("Highest Temperature:", highest)
print("Lowest Temperature:", lowest)
print("Average Temperature:", average)
```

```python
# Activity 2
import numpy as np
board = np.full((3, 3), " ")

board[0, 0] = "X"
board[1, 1] = "O"
board[2, 2] = "X"

def print_board(b):
    print("\nTic-Tac-Toe Board:")
    for row in b:
        print(" | ".join(row))
        print("-" * 9)

print_board(board)
```

```python
# Activity 3
shopping_list = ["milk", "eggs", "bread"]
def show_list():
    print("\nShopping List:")
    for item in shopping_list:
        print("-", item)
    print()

while True:
    print("Options:")
    print("1 - View list")
    print("2 - Add item")
    print("3 - Remove item")
    print("4 - Exit")
    
    choice = input("Choose an option: ").strip()
    
    if choice == "1":
        show_list()
    elif choice == "2":
        item = input("Enter item to add: ").strip().lower()
        shopping_list.append(item)
        print(f"{item} added to the list.")
    elif choice == "3":
        item = input("Enter item to remove: ").strip().lower()
        if item in shopping_list:
            shopping_list.remove(item)
            print(f"{item} removed from the list.")
        else:
            print(f"{item} is not in the list.")
    elif choice == "4":
        print("Exiting program. Final shopping list:")
        show_list()
        break
    else:
        print("Invalid choice. Please enter 1, 2, 3, or 4.")
```

```python
# Activity 4

# Binary Tree
#        50
#       /  \
#     30    70
#    / \    / \
#   20 40  60 80

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

root = Node(50)

root.left = Node(30)
root.right = Node(70)

root.left.left = Node(20)
root.left.right = Node(40)

root.right.left = Node(60)
root.right.right = Node(80)

print("Root:", root.data)
print("Left child of root:", root.left.data)
print("Right child of root:", root.right.data)
print("Children of 30:", root.left.left.data, "and", root.left.right.data)
print("Children of 70:", root.right.left.data, "and", root.right.right.data)
```

```python
# Activity 5
stack = []

print("Type words")
print("Type 'undo' to remove the last word.")
print("Type 'exit' to quit.\n")

while True:
    word = input("Enter a word: ").strip()
    
    if word.lower() == "exit":
        print("Exiting program. Final sentence:")
        print(" ".join(stack))
        break
    elif word.lower() == "undo":
        if stack:  
            removed = stack.pop()
            print(f"Removed '{removed}'. Current sentence: {' '.join(stack)}")
        else:
            print("Nothing to undo.")
    else:
        stack.append(word)
        print("Current sentence:", " ".join(stack))
```

```python
# Activity 6
phone_book = {}

print("Phone Book")
print("Options: add, lookup, delete, view, exit\n")

while True:
    action = input("Choose an option: ").strip().lower()
    
    if action == "add":
        name = input("Enter contact name: ").strip()
        number = input("Enter phone number: ").strip()
        phone_book[name] = number
        print(f"Added {name} with number {number}.")
    
    elif action == "lookup":
        name = input("Enter name to look up: ").strip()
        if name in phone_book:
            print(f"{name}'s number is {phone_book[name]}")
        else:
            print(f"{name} not found in phone book.")
    
    elif action == "delete":
        name = input("Enter name to delete: ").strip()
        if name in phone_book:
            del phone_book[name]
            print(f"{name} deleted from phone book.")
        else:
            print(f"{name} not found in phone book.")
    
    elif action == "view":
        if phone_book:
            print("\nPhone Book Contacts:")
            for name, number in phone_book.items():
                print(f"- {name}: {number}")
            print()
        else:
            print("Phone book is empty.")
    
    elif action == "exit":
        print("Exiting. Final phone book:")
        for name, number in phone_book.items():
            print(f"- {name}: {number}")
        break
    
    else:
        print("Invalid option. Please choose add, lookup, delete, view, or exit.")
```
